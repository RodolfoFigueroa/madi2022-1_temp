Exploraremos la librería NetworkX, que usaremos para trabajar con gráficas en python. La documentación completa es https://networkx.github.io/documentation/stable/_downloads/networkx_reference.pdf

```
import networkx as nx
```

Las clases que usaremos son graph y digraph:
- Graph. Esta clase implementa una gráfica no dirigida, ignora aristas múltiples entre los vértices, y permite que haya aristas que con un mismo vértice en ambos extremos.
- DiGraph. Como el nombre lo dice, esta nos permitirá trabajar con gráficas dirigidas, en caso de ser necesario.

NetworkX cuenta con más clases, pero por el momento con estas dos tendremos para nuestros propósitos.

Esta librería nos permite que los vértices tomen diferentes valores, pueden ser strings, tuplas, enteros y más.

Veamos ahora cómo crear y examinar elementos de una gráfica. Lo primero es definir nuestra gráfica $G$:

```
G = nx.Graph()
```

*Vértices.* Para agregar vértices a la gráfica, se pueden ir agregando de uno en uno o indicar que todos los elementos de cierta lista serán vértices. Recordemos que los vértices pueden tomar diferentes valores, no únicamente enteros. También podemos accesar a una lista con los nodos de $G$.

```
G.add_node(1)
G.add_node('A')
G.add_nodes_from([2, 3, 'B', 'C'])
print(G.nodes)
nx.draw(G, with_labels=True, font_weight='bold', node_color='#71A125')
```

*Aristas.* Las aristas, al igual que los vértices, pueden agregarse de una en una, o por medio de una lista. Además de esto, las aristas declaradas no tienen que ser necesariamente entre vértices ya existentes, en caso de que los vértices utilizados no existan previamente, estos serán agregados a la gráfica. Un atributo que usaremos constantemente es agregar el peso de una arista.

```
G.add_edge(1, 2, weight = 5)
G.add_edge(3, 7)
G.add_edges_from([('A', 'B'), ('C', 'D')])
G.add_weighted_edges_from([(2, 3, 2), (1, 'A', 3)])

print("Lista de los grados por cada vértice:", G.degree, '\n')
print("Grado de un vértice en particular:", G.degree['A'], '\n')
print("Lista de aristas:", G.edges, '\n')
print("Atributo asignado a arista 1,2:", G.edges[1, 2]['weight'], '\n')
print("Atributo asignado a arista 1,2:", G.edges[2, 1], '\n')
print("Atributo asignado a arista 'A','B':", G.edges['A', 'B'], '\n')
print("Lista de vértices adyacentes a 3, con atributos:", G.adj[3], '\n')
nx.draw(G, with_labels=True, font_weight='bold', node_color='#71A125')
```

Ahora, veamos qué pasa con las digráficas, la diferencia es que las aristas tienen dirección, esta librería nos permite conocer tanto el ingrado como el exgrado de los vértices en gráficas de este tipo.

```
DG = nx.DiGraph()

DG.add_nodes_from(['a', 'b', 1, 2, 3, 4])
DG.add_edges_from([(1, 2), ('c', 'b'), (4, 1), ('a', 2), ('a', 'b'), (1, 'a'), (2, 3)])

print("Los vértices de la digráfica son : ", DG.nodes, '\n')
print("Las aristas de la digráfica son : ", DG.edges, '\n')
print("Lista de los ingrados por cada vértice:", DG.in_degree, '\n')
print("Lista de los exgrados por cada vértice:", DG.out_degree, '\n')
print("Ingrado de un vértice en particular:", DG.in_degree[1], '\n')
print("Lista de vértices adyacentes a 1 :", list(DG.successors(1)), '\n')
print("Lista de vértices adyacentes a 1, con atributos :", DG.adj[1], '\n')
print("Lista de aristas que entran a 2 : ", DG.in_edges(2), '\n')
print("Lista de predecesores de 2 : ", list(DG.predecessors(2)))
nx.draw(DG, with_labels=True, font_weight='bold', node_color='#71A125')
```

Además de lo que hemos visto, también se pueden eliminar objetos de la gráfica, por ejemplo si removemos el vértice $1$ de $DG$ se elimina este vértice junto con todas las aristas que lo involucran.

```
DG.remove_node(1)

print("Los vértices ahora son : ", DG.nodes)
print("Las aristas ahora son : ", DG.edges)
nx.draw(DG, with_labels=True, font_weight='bold', node_color='#71A125')
```

Notemos que la celda anterior no se puede volver a ejecutar sin antes definir de nuevo $DG$, esto pues el vértice $1$ ya no estaría en $DG$, y no podemos borrar vértices que no existen en la gráfica.

Ahora, construiremos una gráfica tal que dado un entero $n$, sea un ciclo de longitud $n$. Notemos que un polígono regular de $n$ lados con sus vértices numerados del $0$ al $n-1$ es una representación posible de dicha gráfica.

```
C = nx.Graph()

def CreaC(n) : 
  for i in range (0, n):
    C.add_edge(i, (i+1)%n)

CreaC(10)
print("Los vértices de la gráfica son : ", C.nodes)
print("Las aristas de la gráfica son : ", C.edges)
nx.draw(C, with_labels=True, font_weight='bold', node_color='#71A125')
```
